```python
import pandas as pd
```

```python
#全局配置
DATASET_FOLDER='data'


#movie_lens
ML_100k_CSV=DATASET_FOLDER+"/ml-100k/u.data"
ML_100k_delimiter="\t"

ML_1m_delimiter="::"
ML_1m_MOVIES_CSV=DATASET_FOLDER+"/movies.dat"
```

```python
import nltk
from nltk.tokenize import word_tokenize
def get_tokens(item):
    tokens = word_tokenize(str(item))
    sor=sorted(tokens)
    tk=[]
    for sorted_item in sor:
        #print(sorted_item)
        if(str(sorted_item) in ['(',')',',',"'",'The','the']):
            continue
        # if(str(sorted_item).startswith('19') or str(sorted_item).startswith('20')):
        #     continue
        tk.append(str(sorted_item))
    return ' '.join(tk)
```

```python
movies_title = ['movieid', 'name', 'moviegenres']
movies = pd.read_csv(ML_1m_MOVIES_CSV, sep=ML_1m_delimiter, header=None, names=movies_title,engine='python', encoding='ISO-8859-1')
movies
```

```python
# 创建新的统一名称
movies['new_title']=movies['name'].apply(get_tokens)
movies.head()
```

```python
infos_file=DATASET_FOLDER+'/merged.csv'
infos=pd.read_csv(infos_file)
infos.head()
```

```python
# 创建新的统一名称
infos['new_title']=infos['name'].apply(get_tokens)
infos.head()
```

```python
supply_file=DATASET_FOLDER+'/supply.csv'# 补充数据 infos里面没有的
supply=pd.read_csv(supply_file)
supply.head()
```

```python
# 创建新的统一名称
supply['new_title']=supply['name'].apply(get_tokens)
supply.fillna('',inplace=True)
supply.head()
```

# 合并数据

```python
# 将 df1 和 df2 按照 'name' 列进行合并，保留 df1 中的所有行和 df2 中的匹配行
merged_infos = pd.merge(infos, movies, on='new_title', how='right')
merged_infos=merged_infos[~merged_infos['movieid'].isnull()]
merged_infos.drop({'name_x','new_title','url','time','intro','id'},axis=1,inplace=True)
merged_infos.rename(columns={'name_y':'name','movieid':'id'},inplace=True)
merged_infos = merged_infos.reindex(columns=['id','moviegenres','name','release_time','directors','writers','starts'])
merged_infos
```

```python
merged_supply=pd.merge(movies, supply, on='new_title', how='right')
merged_supply.drop({'name_x','new_title','url','time','intro','create_time','spider_time','title','stars','genres'},axis=1,inplace=True)
merged_supply.rename(columns={'name_y':'name','actors':'starts','release_date':'release_time','movieid':'id'},inplace=True)
merged_supply.reindex(columns=['id','moviegenres','name','release_time','directors','writers','starts'])
merged_supply
```

```python
final_df=pd.concat([merged_supply,merged_infos])
final_df.drop_duplicates(subset=['id'],keep='first',inplace=True)
final_df
```

```python
data=pd.DataFrame(movies)
data.rename(columns={'movieid':'id'},inplace=True)
data.drop({'new_title'},axis=1,inplace=True)
result=pd.concat([final_df,data],axis=0,ignore_index=True)
result.drop_duplicates(subset=['id'],keep='first',inplace=True)
result.rename(columns={'moviegenres':'genres'},inplace=True)
result
```

```python
# 筛选包含 NaN 值的行
mask = result.isna().any(axis=1)
rows_with_nan = result[mask]
rows_with_nan
```

```python
result.fillna('',inplace=True)
```

```python
result = result.sort_values(by='id')
result.to_csv(DATASET_FOLDER+'/result.csv',index=False)
```
